# Pad Sequences

```python
from torch.nn.utils.rnn import pad_sequence

def pad_sequences(sequences, max_len, padding_value=0):
    return [seq + [padding_value]*(max_len - len(seq)) if len(seq) < max_len else seq[:max_len] for seq in sequences]
```

# Create Dataset and DataLoader

```python
class RecommendationDataset(Dataset):
    def __init__(self, sequences, customer_features, product_features, max_seq_len):
        self.sequences = sequences
        self.customer_features = customer_features
        self.product_features = product_features
        self.max_seq_len = max_seq_len

    def __len__(self):
        return len(self.sequences)
    
    def __getitem__(self, idx):
        customer_id, seq, target = self.sequences[idx]

        # Pad sequences
        seq_padded = seq + [0] * (self.max_seq_len - len(seq)) if len(seq) < self.max_seq_len else seq[:self.max_seq_len]
        
        # Get features
        customer_features = self.customer_features[customer_id]
        product_features = self.product_features[target]

        return {
            'customer_features': customer_features,
            'product_features': product_features,
            'sequence': seq_padded,
            'target': target
        }
```

# Building the transformer model

```python
class TransformerRecommendationModel(nn.Module):
    def __init__(self, customer_feat_size, product_feat_size, num_products, embed_dim, nhead, num_layers):
        super(TransformerRecommendationModel, self).__init__()
        
        self.product_embedding = nn.Embedding(num_products, embed_dim)
        self.positional_encoding = nn.Parameter(torch.zeros(1, max_seq_len, embed_dim))
        
        self.transformer_encoder_layer = nn.TransformerEncoderLayer(d_model=embed_dim, nhead=nhead)
        self.transformer_encoder = nn.TransformerEncoder(self.transformer_encoder_layer, num_layers=num_layers)
        
        self.customer_fc = nn.Linear(customer_feat_size, embed_dim)
        self.product_fc = nn.Linear(product_feat_size, embed_dim)
        
        self.output_layer = nn.Linear(embed_dim, num_products)
        
    def forward(self, customer_feat, product_feats, sequence):
        # Embed products in the sequence
        product_embeddings = self.product_embedding(sequence)  # Shape: [batch_size, seq_len, embed_dim]
        
        # Add positional encoding
        product_embeddings += self.positional_encoding[:, :product_embeddings.size(1), :]
        
        # Transform product features
        product_feat_embeds = self.product_fc(product_feats)  # Shape: [batch_size, seq_len, embed_dim]
        
        # Combine embeddings
        embeddings = product_embeddings + product_feat_embeds
        
        # Permute for transformer input
        embeddings = embeddings.permute(1, 0, 2)  # Shape: [seq_len, batch_size, embed_dim]
        
        # Pass through transformer
        transformer_output = self.transformer_encoder(embeddings)  # Shape: [seq_len, batch_size, embed_dim]
        
        # Get the last output
        last_output = transformer_output[-1]  # Shape: [batch_size, embed_dim]
        
        # Transform customer features
        customer_embed = self.customer_fc(customer_feat)  # Shape: [batch_size, embed_dim]
        
        # Combine last output with customer features
        combined = last_output + customer_embed
        
        # Output layer
        output = self.output_layer(combined)  # Shape: [batch_size, num_products]
        
        return output
```
